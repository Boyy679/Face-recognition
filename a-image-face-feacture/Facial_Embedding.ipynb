# Compare the distance between two faces

* The model was trained in a way that faces with a distance of 0.6 or less should be a match. But if you want to be more strict, you can look for a smaller face distance. For example, using a 0.55 cutoff would reduce false positive matches at the risk of more false negatives.

*  Under the hood, we use numpy.linalg.norm (Matrix or vector norm). The function is able to return one of eight different matrix norms, or one of an infinite number of vector norms, depending on the value of the ord parameter.

### Match faces for Obama and biden

```python
image_to_test = face_recognition.load_image_file("data/biden.jpg")
image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]
face_distance = face_recognition.face_distance([obama_face_encoding], image_to_test_encoding)
print(face_distance)
```

### Find multiple pictures for Obama

```python
from IPython.display import Image, display
print("A distance of 0.6 or less means a matching face")
for i in range(6):
    image_file = "data/obama/"+ "0" + str(i) + ".jpg"
    image_to_test = face_recognition.load_image_file(image_file)
    image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]
    face_distance = face_recognition.face_distance([obama_face_encoding], image_to_test_encoding)
    if(face_distance <= 0.6):
        print("Found Boyan in", image_file, "- distance:", face_distance)
        #display(Image(filename=image_file, width=400))
    else:
        print("Cannot find Boyan in", image_file, "- distance:", face_distance)
```

### Find TSNE of faces

```python
import numpy as np
from sklearn.manifold import TSNE

known_encodings = [
    obama_face_encoding,
    biden_face_encoding,
    trump_face_encoding
]
names = ['Obama', 'Biden', 'Trump']
n_classes = 3

reduced = TSNE(n_components=2, random_state=0).fit_transform(known_encodings)

import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline

plt.figure(figsize=(5, 5))
for i in range(n_classes):
    X_label = reduced[i]
    plt.scatter(X_label[0], X_label[1], label = names[i])
plt.legend()
```

### Analyze and visualize the faces by using 3D PCA

```python
known_encodings = [
    obama_face_encoding,
    biden_face_encoding,
    trump_face_encoding
]
names = ['Obama', 'Biden', 'Trump']
n_classes = 4

for i in range(10):
    image_file = 'data/oboma&trump&biden/'+"0" + str(i) + ".jpg"
    image_to_test = face_recognition.load_image_file(image_file)
    image_to_test_encoding = face_recognition.face_encodings(image_to_test)[0]
    known_encodings.append(image_to_test_encoding)
    print("Processing file:", image_file)
```

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d import proj3d

import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False

X_obama = [known_encodings[0]]
X_trump = [known_encodings[1]]
X_biden = [known_encodings[2]]
for i in range(0, 3):
    X_obama.append(known_encodings[i])  
for i in range(3, 8):
    X_trump.append(known_encodings[i])  
for i in range(8, 10):
    X_biden.append(known_encodings[i]) 
```

```python
pca = PCA(n_components=3).fit(known_encodings)
Xd_obama = pca.transform(X_obama)
Xd_trump = pca.transform(X_trump)
Xd_biden = pca.transform(X_biden)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
plt.rcParams['legend.fontsize'] = 10   
ax.plot(Xd_obama[:,0], Xd_obama[:,1], Xd_obama[:,2],
        'o', markersize=8, color='red', alpha=0.5, label='Obama')
ax.plot(Xd_trump[:,0], Xd_trump[:,1], Xd_trump[:,2],
        'o', markersize=8, color='purple', alpha=0.5, label='Trump')
ax.plot(Xd_biden[:,0], Xd_biden[:,1], Xd_biden[:,2],
        'o', markersize=8, color='green', alpha=0.5, label='Biden')

plt.title('3D PCA Face Embedding')
ax.legend(loc='upper right')

plt.show()
```

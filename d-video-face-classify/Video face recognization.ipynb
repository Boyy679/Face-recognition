```python
import cv2
import face_recognition
```

# Assume input video test.mp4

If you have known face encodings, you can add here:

```python
features = []
frame_num = 1

cap = cv2.VideoCapture('data/test.mp4')
while(True):
    ret, frame = cap.read()
    if not ret:
        break
    img = frame

    face_locations = face_recognition.face_locations(img)
    if(len(face_locations) > 0):
        face_encodings = face_recognition.face_encodings(img, face_locations)[0]
        features.append(face_encodings)
        print("Processing frame: {}".format(frame_num))
        print("Number of faces detected: {}".format(len(face_locations)))
    frame_num = frame_num + 1

cap.release()
```

# t-SNE transform

```python
from sklearn.manifold import TSNE
reduced = TSNE(n_components=2, random_state=0).fit_transform(features)
```

# Plot and see the face encoding distribution

```python
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus']=False
%matplotlib inline

plt.figure(figsize=(5, 5))
plt.scatter(x = reduced[:,0], y=reduced[:,1])
plt.show()
```
